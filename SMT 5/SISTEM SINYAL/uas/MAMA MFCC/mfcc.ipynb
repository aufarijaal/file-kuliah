```python
try:
  import warnings
  import os
  warnings.filterwarnings('ignore')
  import numpy as np
  import matplotlib.pyplot as plt
  import matplotlib.style as ms
  ms.use('seaborn-muted')
  %matplotlib inline

  import librosa
  import librosa.display
  import IPython.display

  print("import sukses")
except:
  print("""ada library yang gada.
install dulu pakai pip install -r library_yang_dibutuhkan.txt
  """)
```

```python

def convert_to_mel_spectogram(suara, ambil_lognya_saja = False):
  y, sr = librosa.load(suara)
  S = librosa.feature.melspectrogram(y, sr=sr, n_mels=128)
  log_S = librosa.logamplitude(S, ref_power=np.max)
  if ambil_lognya_saja == True:
    return log_S
  plt.figure(figsize=(12,4))
  librosa.display.specshow(log_S, sr=sr, x_axis='time', y_axis='mel')
  plt.title('Mel Spectogram dari ' + os.path.basename(suara))
  plt.colorbar(format='%+02.0f dB')
  plt.tight_layout()

convert_to_mel_spectogram("../Suara/Fix/0_mama_umi.wav")
convert_to_mel_spectogram("../Suara/Fix/1_mama_yuli.wav")
convert_to_mel_spectogram("../Suara/Fix/2_mama_rezi.wav")
convert_to_mel_spectogram("../Suara/Fix/3_mama_andi.wav")
convert_to_mel_spectogram("../Suara/Fix/4_mama_bocah1.wav")
convert_to_mel_spectogram("../Suara/Fix/5_mama_bocah2.wav")
convert_to_mel_spectogram("../Suara/Fix/6_mama_aku.wav")
```

```python
def display_mfcc(path):
  mfcc = librosa.feature.mfcc(S=convert_to_mel_spectogram(path, True), n_mfcc=13)

  plt.figure(figsize=(12, 6))

  plt.subplot(3,1,1)
  librosa.display.specshow(mfcc)
  plt.ylabel('MFCC')
  plt.title("hasil gambar dari " + os.path.basename(path))
  plt.colorbar()

display_mfcc("../Suara/Fix/0_mama_umi.wav")
display_mfcc("../Suara/Fix/1_mama_yuli.wav")
display_mfcc("../Suara/Fix/2_mama_rezi.wav")
display_mfcc("../Suara/Fix/3_mama_andi.wav")
display_mfcc("../Suara/Fix/4_mama_bocah1.wav")
display_mfcc("../Suara/Fix/5_mama_bocah2.wav")
display_mfcc("../Suara/Fix/6_mama_aku.wav")
```
