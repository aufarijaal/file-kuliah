```python
try:
  import warnings
  import os
  warnings.filterwarnings('ignore')
  import numpy as np
  import matplotlib.pyplot as plt
  import matplotlib.style as ms
  ms.use('seaborn-muted')
  %matplotlib inline

  import librosa
  import librosa.display
  import IPython.display

  print("import sukses")
except:
  print("""ada library yang gada.
install dulu pakai pip install -r library_yang_dibutuhkan.txt
  """)
```

```python
def stft(file):
  fig = plt.figure(figsize=(20, 64))
  y, sr = librosa.load(file)
  plt.subplot(10, 1,1)
  D = librosa.logamplitude(np.abs(librosa.stft(y)) ** 2, ref_power=np.max)
  librosa.display.specshow(D, sr=sr, y_axis='log')
  plt.title("STFT dari " + os.path.basename(file))
  plt.tight_layout(pad=10)
  plt.show()

stft("../Suara/Fix/0_mama_umi.wav")
stft("../Suara/Fix/1_mama_yuli.wav")
stft("../Suara/Fix/2_mama_rezi.wav")
stft("../Suara/Fix/3_mama_andi.wav")
stft("../Suara/Fix/4_mama_bocah1.wav")
stft("../Suara/Fix/5_mama_bocah2.wav")
stft("../Suara/Fix/6_mama_aku.wav")
```
