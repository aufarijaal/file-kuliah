```python
import os
import sys
sys.path.append('library\\thinkdsp')
try:
  import thinkdsp
except ModuleNotFoundError as er:
  if 'scipy' in str(er):
    print('install module scipy dulu gan\npip install scipy')
  elif 'matplotlib' in str(er):
    print('install module matplotlib dulu gan\npip install matplotlib')
```

membuat audio dari file `data/gitar_kunci_g.wav`

```python
wave = thinkdsp.read_wave('data/gitar_kunci_g.wav')
wave.make_audio()
```

melakukan segmentasi pada audio. memilih bagian mana yang akan diambil. lalu mendekorasikannya supaya enak dilihat

## GRAFIK AMPLITUDO TERHADAP WAKTU

```python
start = 1.4
duration = 5
segment = wave.segment(start, duration)
segment.plot()
thinkdsp.decorate(xlabel='Waktu (d)')
thinkdsp.decorate(ylabel='Amplitudo')
```

## GRAFIK PLOT SPEKTRUM FREKUENSI

```python
spectrum = segment.make_spectrum()
spectrum.plot()
thinkdsp.decorate(xlabel='Frekuensi (Hz)')
thinkdsp.decorate(ylabel='Amplitudo')

print('frekuensi tertinggi adalah ')
```
